```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
data=pd.read_csv("Churn_Modelling.csv")
data.head()
```

```python
data.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
#eEncoding the categorical data
label_encoder_gender=LabelEncoder()
data['Gender']=label_encoder_gender.fit_transform(data['Gender'])
```

```python
data
```

```python
#one hot encoding for geographical 
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo=OneHotEncoder()
geo_encoder=onehot_encoder_geo.fit_transform(data[['Geography']])
geo_encoder
```

```python
geo_encoder.toarray
```

```python
onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoder_df=pd.DataFrame(geo_encoder.toarray(),columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
```

```python
geo_encoder_df
```

```python
data=pd.concat([data.drop('Geography',axis=1),geo_encoder_df],axis=1)
```

```python
data.head()
```

```python
data=data.drop('Surname',axis=1)
```

```python
data.head()
```

```python
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)
with open('label_encoder_geo.pkl','wb') as file:
    pickle.dump(onehot_encoder_geo,file)
```

```python
#divide the dataset to dependent and independent vfeatures
X=data.drop('Exited',axis=1)
y=data['Exited'] 

#split the data to training  and testing sets
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=22)


```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

## ANN Implementation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
model=Sequential([
    Dense(64,activation='relu',input_shape=(X_train.shape[1],)), #hidden layer 1 connected with input layer
    Dense(32,activation='relu'),  # hidden layer 2
    Dense(1,activation='sigmoid') # output layer
]
)
```

```python
model.summary()
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.1)
loss=tensorflow.keras.losses.BinaryCrossentropy()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
## setup the tensorboard
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
log_dir="logs/fit" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python

early_stopping_callback=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
# Training the model
history=model.fit(
    X_train,y_train,validation_data=(X_test,y_test),epochs=100,
    callbacks=[tensorflow_callback,early_stopping_callback]
    )
                               
```

```python
model.save('model.h5')
```
